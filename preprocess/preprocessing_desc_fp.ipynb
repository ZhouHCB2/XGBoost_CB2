```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
fingerprint_list = ['MorganFP', 'RDkitFP', 'AtomPairFP', 'TorsionFP', 'AvalonFP', 
'EstateFP', 'MACCSFP', 'PharmacoErGFP', 'PharmacoPFP', 'PubChemFP', 'MHFP6', 'MAP4']
descriptor_list = ['Property', 'Constitution', 'Autocorr', 'Fragment', 'Charge', 
'Estate', 'MOE', 'Connectivity', 'Topology', 'Kappa', 'Path', 'Matrix', 'InfoContent']
```

```python
df=pd.DataFrame()
```

```python
for fp in descriptor_list:
    fp_file=f'ChblReg6227_{fp}.csv';#print(fp_file)
    tdf=pd.read_csv(fp_file)
    df=pd.concat([df,tdf],axis=1)
```

```python
print(len(df));print(len(df.columns))
```

```python
df.to_csv('all_desc.csv',index=False)
```
