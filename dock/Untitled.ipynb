```python
import pandas as pd
```

```python
df = pd.read_csv('5ZTY_docked.csv')
df1 = pd.read_csv('6KPF_docked.csv')
```

```python
y_pred = df['docking_score']
y_true = df['act4']

y1_pred = df1['docking_score']
y1_true = df1['act4']
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc   
```

```python
fpr1, tpr1, _ = roc_curve(y_true,y_pred)
fpr2, tpr2, _ = roc_curve(y1_true,y1_pred)
plt.figure()
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr1, tpr1, label='5ZTY(AUC=0.663)',color='red')
plt.plot(fpr2, tpr2, label='6KPF(AUC=0.512)',color='yellow')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('cb2_random_r ROC curve')
plt.legend(loc='best')
plt.rcParams['figure.figsize'] = (5.0, 5.0)
plt.show()
plt.savefig('docked_ROC.tiff',dpi=300)
```
